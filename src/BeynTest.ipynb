Now the goal is to compute the complex contour integral 

$A_0 = \frac{1}{2\pi i}\int\limits_{\Gamma} T(z)^{-1} \hat{V} dz $

$A_1 = \frac{1}{2\pi i}\int\limits_{\Gamma} z T(z)^{-1} \hat{V} dz $

this satisfies $A_0,A_1 \in \mathcal{C}^{m,l}$ 

Using the trapezoid sum, the integral can be approximated as: 

$z = \phi(t)$

$A_0 = \frac{1}{2\pi i}\int\limits_{0}^{2\pi} T(\phi(t))^{-1} \hat{V}\phi\prime(t) dt $

$A_0 \approx A_{0,N}= \frac{1}{i N}\sum\limits_{j=0}^{N-1} T(\phi(t_j))^{-1} \hat{V}\phi\prime(t_j) dz $
